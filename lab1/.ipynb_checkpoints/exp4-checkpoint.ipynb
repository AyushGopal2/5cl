```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
w0 = 18 # px
w = np.array([63, 71, 60, 75, 117]) # px
dw = 2 # px

s = np.array([25.7,33.4,28.6,33,49.9]) # cm
ds = 0.1 # cm

d0 = 10.5 # cm
dd0 = 0.5 # cm

fobj = 17.3 # cm
dfobj = 0.05 # cm

fep = 4.8 # cm
dfep = 0.12 # cm

mag = w/w0
dmag = dw/w0
model = lambda s1: s1*d0/fobj/fep
dmodel = lambda s1: np.sqrt(
(fep * fobj * s1 * dd0)**2
+ (fep * fobj * d0 * ds)**2    
+ (fep * s1 * d0 * dfobj)**2    
+ (fobj * s1 * d0 * dfep)**2    
) / (fep * fobj)**2

res = mag - model(s)
norm_res = res/dmodel(s)
dres = dmag/dmodel(s)
```

```python
chi2 = np.sum(norm_res**2)
rchi2 = chi2 / 4
chi2, rchi2
```

```python
fig, (ax_main, ax_res) = plt.subplots(
        2, 1, sharex=True, 
    figsize=(6, 4), dpi=200, 
    gridspec_kw={'height_ratios': [5, 3]}
)

plt.sca(ax_main)
plt.title("Magnification of a compound microscope")

plt.errorbar(
    s, mag, yerr = dmag,
    ecolor="blue", c="black",
    label='Measurements',
    linewidth = 0, elinewidth = 1, markersize = 8, marker='.'
)


s_range = np.linspace(20, 55)
plt.plot(s_range, model(s_range),
            label=f"Paraxial model", color='orange'
)
plt.gca().fill_between(
    s_range, 
    model(s_range) + dmodel(s_range), model(s_range) - dmodel(s_range),
    color='orange', alpha=0.2,
    label=f"Model uncertainty (1)"
)


plt.ylabel("Angular magnification")
plt.grid()
plt.legend(loc='upper left')

plt.sca(ax_res)
plt.errorbar(
    s, norm_res, yerr = dres,
    ecolor="blue", c="black",
    label=f'$\\tilde\chi^2 = {rchi2:.2f}$',
    linewidth = 0, elinewidth = 1, markersize = 8, marker='.'
)
plt.gca().fill_between(
    s_range, 
    -np.ones_like(s_range),
    np.ones_like(s_range),
    color='orange', alpha=0.2
)
plt.axhline(0, color='orange')

plt.xlabel("Focal point separation (cm)")
plt.ylabel("Residual ()")
plt.grid()
plt.legend(loc='lower right')
plt.ylim(-2, 2)

plt.xlim(np.min(s_range), np.max(s_range))

plt.tight_layout()
plt.savefig('exp4_0.png')
```
