```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
plt.rcParams['font.family'] = 'FreeSerif'
plt.rcParams['mathtext.fontset'] = 'cm'
```

```python
metals = ['al', 'brass', 'cu']
colors = { 'al': ['olive', 'slategray'], 'brass': ['seagreen', 'steelblue'],  'cu': ['coral', 'crimson']}
labels = { 'al': 'Aluminum', 'brass': 'Brass',  'cu': 'Copper'}
ls = { 'al': '-', 'brass': '--',  'cu': '-.'}

file = lambda metal, n: f'data/exp4_{metal}_{int(n)}.csv'

N_convolve = 2**7
gauss = np.exp(-np.linspace(-2,2, N_convolve)**2)
conv = lambda d: np.convolve(d, gauss/np.sum(gauss), mode='valid')

data = dict()
for m in metals:
    data[m] = []
    for n in [1, 2]:
        [t, T]  = np.genfromtxt(file(m,n), delimiter=',')[1:].T
        i = np.argmax(T)
        tt, TT = t[i:] - t[i], T[i:]
        grad = np.gradient(TT, tt)
        j = np.argmin(conv(grad))
        ttt, TTT, dTdt = tt[j:] - tt[j], TT[j:], grad[j:]
        data[m].append([ttt, TTT])
```

```python
fig, ax = plt.subplots(figsize=(5,2), dpi=200)
s = []
for m in metals:
    for (i, (t, T)) in enumerate(data[m]):
        print(np.mean(np.diff(t))*N_convolve)
        plt.plot(conv(t), T[N_convolve//2:1-N_convolve//2] - conv(T), color='black', alpha=0.5)
        s.append(np.std(T[N_convolve//2:1-N_convolve//2] - conv(T)))

plt.xlabel('Time (s)')
plt.grid()
plt.ylabel('Residuals (°C)')
plt.ylim(-0.5, 0.5)
noise = np.mean(s)
np.mean(s), np.std(s)
```

```python
# model = lambda t, t0, T0, T1, T2: (
#     T0 * np.exp(-t/t0) + T1 * np.exp(-4*t/t0) + T2 * np.exp(-9*t/t0)
# )
model = lambda t, t0, T0, T1: (
    T0 + T1 * np.exp(-t/t0)
)

fit = dict()
for m in metals:
    fit[m] = []
    for (i, [t, T]) in enumerate(data[m]):
        xs, cov = sp.optimize.curve_fit(
            model, t, T, 
            p0=[100, 0, 15],
            bounds=([0, 0, 10], [3000, 20, 50])
        )
        dxs = np.sqrt(np.diagonal(cov))
        print(m, xs, dxs)
        out = [(x, dxs[i]) for (i, x) in enumerate(xs)]
        fit[m].append(out)
    fit[m] = np.array(fit[m])
```

```python
T0 = [ fit[m][0,1,0] for m in metals ]
np.mean(T0), np.std(T0)
```

```python
L = 185 # mm
dL = 5

cs = {
    'al': (0.900, 0.001),
    'brass': (0.380, 0.001),
    'cu': (0.386, 0.001)
} # specific heat capacity and error, J / g K

rhos = {
    'al': (2.703, 0.11),
    'brass': (8.824, 0.11),
    'cu': (8.387, 0.11),
} # mass density and error, g / mL

print('alloy, diffusion timescale (s), thermal diffusivity (mm^2/s), thermal conductivity (J / m s K)')
for (k, m) in enumerate(metals):
    
    ts, dts = fit[m][:,0].T
    tvar = 1/np.sum(1/dts/dts)
    tmean = np.sum(ts/dts/dts)*tvar
    # dt = np.sqrt(tvar)
    dt = np.std(ts)
    

    c, dc = cs[m]
    rho, drho = rhos[m]
    alpha = (3*L/np.pi)**2 /tmean
    dalpha = (3/np.pi)**2*np.sqrt((dt*L*L/tmean/tmean)**2 + (2*L*dL/tmean)**2+(0.5/3)**2)
    k = c * alpha * rho / 1e6 * 1e6
    dk = k * np.sqrt((dc/c)**2 + (dalpha/alpha)**2 + (drho/rho)**2)
    
    print(m, (tmean, dt), (alpha, dalpha), (k, dk) )
```

```python
fig, (axs, res_axs) = plt.subplots(
    nrows=2, ncols=3, 
    figsize=(8, 5), dpi=200, sharey='row', sharex=True,
    gridspec_kw={'height_ratios': [4, 3]}
)
axs[0].set_ylabel('Temperature (°C)')
res_axs[0].set_ylabel('Residuals (°C)')
res_axs[1].set_xlabel('Time (s)')

for (k, m) in enumerate(metals):
    plt.sca(axs[k])
    plt.title(labels[m])
    
    for (i, (t, T)) in enumerate(data[m]):
        ls = ('dashed' if i else 'solid')
        c = colors[m][i]
        
        (t0, dt0), (T0, dT0), (T1, dT1) = fit[m][i]

        plt.sca(axs[k])
        plt.plot(t, T, 
                 color=c, ls=ls,
                 label=f'Trial {i+1}:\n $t_\\mathrm{{T}} = {t0:.0f}\pm{dt0:.0f}$ s,\n $T_\\mathrm{{bath}} = {T0:.1f}\pm{dT0:.1f} ^\circ$C')
        
        mean = model(t, *[x for (x, dx) in fit[m][i]])
        samples = np.array([
            model(t, *[np.random.normal(x, dx) for (x, dx) in fit[m][i]])
            for j in range(50)
        ])
        std = np.sqrt(np.std(samples, axis=0)**2 + noise**2)
        res = T - mean
        rchi2 = np.mean((res/std)**2)

        ## model

        plt.sca(axs[k])
        t_range = np.linspace(np.min(t), 600, 100)
        mean_range = model(t_range, *[x for (x, dx) in fit[m][i]])
        sample_range = np.array([
            model(t_range, *[np.random.normal(x, dx) for (x, dx) in fit[m][i]])
            for j in range(50)
        ])
        std_range = np.sqrt(np.std(sample_range, axis=0)**2 + noise**2)
        plt.fill_between(t_range,
                         mean_range - std_range, 
                         mean_range + std_range,
                         color=c, alpha=0.5)
        
        plt.sca(res_axs[k])
        plt.plot(t, res, color=c, ls=ls)
        plt.fill_between(t_range, -std_range, std_range, color=c, alpha=0.3, ls=ls,
                        label=f'Fit {i+1}: $\\tilde\chi^2 = {rchi2:.3f}$')

    plt.sca(axs[k])
    plt.legend()
    plt.grid()
    
    plt.xlim(0, 500)
    plt.ylim(10, 30)

    plt.sca(res_axs[k])
    plt.grid()
    plt.ylim(-0.3,0.3)
    plt.legend(loc='lower right')
    
plt.tight_layout()
plt.show()
```
