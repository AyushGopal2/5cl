```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
theta = np.radians(np.array([0,15,30,45,60,75,90,105,120,135,150,165,180])) # rad
dtheta = np.radians(5) # rad
l = np.array([68.08,65.81,49.32,40.55,28.56,18.13,13.13,14.07,23.48,33.7,49.88,57.12,64.08]) #lux
dl = np.array([0.86,0.402,0.132,1.45,0.72,0.254,0.0918,0.1375,0.4444,1.43,1.169,1.54,2.79]) # lux
```

```python
model = lambda th, lmax, l0, th0: lmax * np.cos(th - th0)**2 + l0
[lmax, l0, th0], cov = sp.optimize.curve_fit(
    model, 
    theta, l,
    p0 = [52, 12, 0]
)
dlmax = np.sqrt(cov[0,0])
dl0 = np.sqrt(cov[1,1])
dth0 = np.sqrt(cov[2,2])
lmax, l0, th0, cov
```

```python
space = np.linspace(-1, np.pi + 1, 100)
l_model = model(space, lmax, l0, th0)
sigma_model = 0.5*(model(space, lmax+dlmax, l0+dl0, th0) - model(space, lmax-dlmax, l0-dl0, th0))
res = l - model(theta, lmax, l0, th0)
sigma = 0.5*(model(theta, lmax+dlmax, l0+dl0, th0) - model(theta, lmax-dlmax, l0-dl0, th0))
dres = dl/sigma
normres = res/sigma
```

```python
(lmax, dlmax), (l0, dl0), (np.degrees(th0), np.degrees(dth0))
```

```python
chi2 = np.sum(res**2/sigma)
rchi2 = chi2 / (theta.size + 2)
chi2, rchi2
```

```python
fig, (ax_main, ax_res) = plt.subplots(
        2, 1, sharex=True, 
    figsize=(6, 6), dpi=200, 
    gridspec_kw={'height_ratios': [5, 2]}
)

plt.sca(ax_main)

plt.plot(np.degrees(space), l_model, c='orange',
    label="$I(\\theta) = I_0\cos^2(\\theta - \\theta_0) + I_{bg}$"
)
ax_main.fill_between(
    np.degrees(space), 
    l_model - sigma_model,
    l_model + sigma_model,
    color='orange', alpha=0.2,
    label="$\pm1\sigma$"
)

plt.errorbar(
    np.degrees(theta), l, yerr = dl, xerr = np.degrees(dtheta),
    ecolor="blue", c="black",
    label='Measurements',
    linewidth = 0, elinewidth = 1, markersize = 4, marker='.',
)
plt.ylabel("Intensity (lux)")
plt.grid()
plt.legend()

plt.sca(ax_res)
plt.errorbar(
    np.degrees(theta), normres, yerr = dres, xerr = np.degrees(dtheta),
    ecolor="blue", c="black",
    label=f'$\\tilde\chi^2 = {rchi2:.2f}$',
    linewidth = 0, elinewidth = 1, markersize = 4, marker='.'
)
ax_res.fill_between(
    np.degrees(space), 
    -np.ones_like(space),
    np.ones_like(space),
    color='orange', alpha=0.2
)
plt.axhline(0, color='orange')

plt.plot()
plt.ylabel("Residuals")
plt.xlabel("Angle (deg)")
plt.xlim(-5, 185)
plt.ylim(-2.5, 2.5)
plt.grid()
plt.legend()
plt.tight_layout()
```
