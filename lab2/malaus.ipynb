```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
theta = np.radians(np.array([0,15,30,45,60,75,90,105,120,135,150,165,180])) # rad
l = np.array([68.08,65.81,49.32,40.55,28.56,18.13,13.13,14.07,23.48,33.7,49.88,57.12,64.08]) #lux
dl = np.array([0.86,0.402,0.132,1.45,0.72,0.254,0.0918,0.1375,0.4444,1.43,1.169,1.54,2.79]) # lux
```

```python
model = lambda th, lmax, l0: lmax * np.cos(th)**2 + l0
[lmax, l0], cov = sp.optimize.curve_fit(
    model, 
    theta, l
    # p0 = [24, 1e4]
)
lmax, l0, cov
```

```python
space = np.linspace(0, np.pi)
plt.scatter(theta, l)
plt.plot(space, model(space, lmax, l0))
plt.ylabel("Intensity (lux)")
plt.xlabel("Angle (rad)")
plt.grid()
```
