```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
x = np.array([
    24.3, 26.5, 28.0, 32.0, 36.0, 40.0
]) # cm
```

```python
l = np.array([
    3.06E+04, 1.44E+04, 7.59E+03, 2.80E+03, 1.55E+03, 9.49E+02
]) # lux
```

```python
model = lambda x, x0, l0: l0 * (x-x0)**(-2)
[x0, l0], cov = sp.optimize.curve_fit(
    model, 
    x, l,
    p0 = [24, 1e4]
)
x0, l0, cov
```

```python
y = x - x0
space = np.linspace(np.min(y), np.max(y))
plt.scatter(y, l)
plt.plot(space, model(space, 0, l0))
plt.ylabel("Intensity (lux)")
plt.xlabel("Distance (cm)")
plt.grid()
```
