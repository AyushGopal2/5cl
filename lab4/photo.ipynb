```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

## Monte Carlo

```python
# all dimensionful quantities in meV

def montecarlo(
    V, # voltage bias
    Phi=2000, # work function (guess)
    f=2100, # photon frequency (600 nm)
    T=26, # temperature (300 K)
    fermi=300, # fermi energy (Cesium)
    power=0.9,
    LED=False, 
    N=10000, # number of samples
):

    fspread = 70 # LED spread
    gauss = lambda df: np.exp(-(df/fspread)**2/2)
    
    m = 5.1E8 # electron mass
    maxwell = lambda v, T=26: v*v * np.exp(-m*v*v/(2*T))
    
    norm = lambda a: a/np.sum(a)
    
    dfspace = np.linspace(-5*fspread, 5*fspread, N) # frequency
    vspace = np.linspace(0, 10*np.sqrt(T/m), N) # speed
    espace = np.linspace(0, fermi, N) # band
    
    df = np.random.choice(dfspace, size=N, p=norm(gauss(dfspace))) if LED else 0
    v = np.random.choice(vspace, size=N, p=norm(maxwell(vspace, T=T)))
    e = np.random.choice(espace, size=N, p=norm(espace**power))
            
    K = 0.5* m*v**2 + f + df - Phi - e # kinetic energy

    KK, VV = np.meshgrid(K, V)

    I = np.mean(np.logical_and(KK > 0, KK > -VV), axis=1)

    if fermi > 0.1:
        Vstop = Phi - f
        V0 = min(0, Phi + fermi - f) - 200
        I[V > V0] = np.max(I[V < V0]) + (V[V>V0]-V0) * (np.max(I[V < V0]) - np.max(I[V < V0-10]))/10

    return I
    # return I / np.min(I[V>0])
```

```python
fig, ax = plt.subplots(figsize=(6, 3), dpi=200)

# Vspace = np.linspace(-2000, 100, 200)
Vspace = np.linspace(-2000, 100, 200)
plt.xlim(np.min(Vspace), np.max(Vspace))
plt.plot(Vspace, montecarlo(Vspace, Phi=1500, f=2100, T=0.001, fermi=0.01), label='$T=0$, $E_F = 0$')
plt.plot(Vspace, montecarlo(Vspace, Phi=1500, f=2100, T=0.001), label='$T=0$, $E_F = 1.6$ eV')
plt.plot(Vspace, montecarlo(Vspace, Phi=1500, f=2100), label='$T=300$ K, $E_F = 1.6$ eV')
plt.xlabel('Voltage (mV)')
plt.ylabel('Electrons at anode per photon')
plt.ylim(0,1)
plt.xlim(-700, -200)
plt.legend(loc='upper left')
plt.grid()
```

```python
params = dict()
models = dict()
for k in V:
    model = lambda Vspace, I0, Phi, fermi, power: I0 * montecarlo(
        Vspace, Phi=Phi, f=f[k], fermi=fermi, power=power, LED=k.isupper(), N=1000
    )
    V0 = np.max(V[k][I[k] < 10])
    [I0, Phi, fermi, power], cov = sp.optimize.curve_fit(
        model, V[k], I[k],
        p0 = [np.min(I[k][V[k] > 0]), f[k] + V0, 1600, 0.8],
    )
    models[k] = lambda Vspace: model(Vspace, I0, Phi, fermi, power)
    params[k] = (I0, Phi, fermi, power, cov)
    print(I0, Phi, fermi, power)
```
