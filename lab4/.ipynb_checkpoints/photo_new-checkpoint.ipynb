```python
# lowercase = laser
# uppercase = LED
I = {
    'B': [0,3,9,12,21,49,74,149,250,387,552,718,847,903,942,953,974,1020,1105,1278,1475,1747],
    'G': [356,265,189,89,50,14,7,7,1,1,30,18,76,107,196,150,569,674,738,833,898,1033],
    'R': [25,43,5,3,3,1,1,0,20,14,9,7,1,61,67,80,94],
    'g': [545,52,207,7,70,12,218,458,134,363,296,161,29,101,20,10,3,3,1,569,596,663,727,814,894],
    'b': [2126,65,21,0,1790,1475,1087,627,325,78,45,207,269,129,391,658,998,2408,2660,2904,3075,2719,2189],
    'r': [54,10,23,34,50,20,7,7,1,1,7,3,7,0,1,20,96,65,76,132,165,207,229,250,320],
} # nano amps

V = {
    'B': [-1000,-956,-902,-873,-814,-724,-675,-558,-460,-357,-248,-150,-75,-45,-25,-19,-6,0,40,125,216,340],
    'G': [-121,-191,-259,-366,-433,-557,-576,-617,-663,-681,-492,-519,-395,-352,-255,-301,20,90,126,183,224,305],
    'R': [-66,0,-150,-188,-200,-213,-220,-269,-104,-126,-163,-198,-243,55,77,129,189],
    'g': [-3,-445,-267,-589,-418,-556,-253,-64,-338,-134,-185,-308,-506,-373,-534,-583,-625,-648,-679,11,33,76,116,174,225],
    'b': [1,-1070,-1200,-1370,-112,-226,-387,-610,-795,-1050,-1120,-892,-836,-977,-748,-595,-427,98,181,258,311,201,30],
    'r': [0,-108,-54,-34,-13,-75,-117,-175,-230,-198,-118,-158,-136,-288,-253,-89,64,14,30,112,154,199,227,259,327]
} # milli volts

for o in [I, V]: # array conversion
    for k in o:
        o[k] = np.array(o[k], dtype=float)


lam = {
    'R': (1.5*539 + 0.5*601)/2,
    'G': (1.5*437 + 0.5*593)/2,
    'B': (1.5*391 + 0.5*501)/2,
    'r': 635,
    'g': 532,
    'b': 405
} # nm
f = dict()
for k in lam:
    f[k] = 1.24E6/lam[k] # meV


label = {
    'R': 'red LED',
    'G': 'green LED',
    'B': 'blue LED',
    'r': 'red laser',
    'g': 'green laser',
    'b': 'violet laser'
} # nm
```

```python
params = dict()
Vs = dict()
# model = lambda Vspace, V0, m: np.clip(m * (Vspace - V0), a_min=0, a_max=None)
model = lambda Vspace, V0, a, b, p: a * (np.clip(Vspace - V0, a_min=0, a_max=None))**(1+p) / (b + np.clip(Vspace - V0, a_min=0, a_max=None)**p)
for k in V:
    [V0, m, b, p], cov = sp.optimize.curve_fit(
        model, V[k], I[k], 
        p0=[np.max(V[k][I[k] < 10]), 0.5, 1e6, 2],
        bounds=([-2000, 0, 0, 1], [0, 100, 1e7, 4]),
    )
    params[k] = (V0, m, b, p)
    Vs[k] = (V0, np.sqrt(cov[0][0])+50)
    print(V0, m, b, p)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=200)

Vspace = np.linspace(-1500, 300, 100)
for k in label:
    col = wavelength_to_rgb(lam[k]) * 0.8
    plt.scatter(V[k], I[k], color=col, s=20, marker=('s' if k.isupper() else '^'), 
                label=f'{label[k]}: ${Vs[k][0]:.0f} \pm {Vs[k][1]:.0f}$ mV')
    plt.plot(Vspace, model(Vspace, *params[k]), color=col)
plt.ylim(0, 500)
plt.xlim(-1500, 300)
plt.grid()
plt.xlabel('Voltage (mV)')
plt.ylabel('Current (nA)')
plt.legend()
```

```python
fig, (ax, ax_res) = plt.subplots(
        2, 1, sharex=True, 
    figsize=(6, 6), dpi=200, 
    gridspec_kw={'height_ratios': [5, 2]}
)

plt.sca(ax)

planck = lambda f, h, f0: h * (f - f0)

ledf = np.array([3e8/(lam[k]*1e-9) for k in label if k.isupper()])
ledv = np.array([-Vs[k][0]/1000 for k in label if k.isupper()])
leddv = np.array([-Vs[k][1]/1000 for k in label if k.isupper()])

[ledh, ledf0], ledcov = sp.optimize.curve_fit(
    planck, ledf, ledv, 
    p0=[1e-14, 1e12],
)
leddh = np.sqrt(ledcov[0,0])
leddf = np.sqrt(ledcov[1,1])
[lash, lasf0], lascov = sp.optimize.curve_fit(
    planck, lasf, lasv, 
    p0=[1e-14, 1e12],
)
lasdh = np.sqrt(lascov[0,0])
lasdf = np.sqrt(lascov[1,1])


print(ledh, ledf0)


lasf = np.array([3e8/(lam[k]*1e-9) for k in label if k.islower()])
lasv = np.array([-Vs[k][0]/1000 for k in label if k.islower()])
lasdv = np.array([-Vs[k][1]/1000 for k in label if k.islower()])

fspace = np.linspace(0, 1e15, 10)
plt.errorbar(
    ledf/1e12, ledv, yerr = leddv,
    ecolor="gray", color='blue', label='LED',
    linewidth = 0, elinewidth = 1, markersize = 4, marker='s'
)

led0 = planck(ledf, ledh, ledf0)
led1 = planck(ledf, ledh - leddh, ledf0 + leddf)
led2 = planck(ledf, ledh + leddh, ledf0 - leddf)
plt.plot(fspace/1e12, planck(fspace, ledh, ledf0), color='blue', lw=1,
        label=f'$f_0 = {ledf0*1e-12:.0f} \pm {leddf*1e-12:.0f}$ THz,\n $h = ({ledh*1e15:.1f} \pm {leddh*1e15:.1f})\cdot10^{{-15}}$ eV/Hz')
ax.fill_between(
    fspace/1e12, 
    planck(fspace, ledh - leddh, ledf0 + leddf),
    planck(fspace, ledh + leddh, ledf0 - leddf),
    color='blue', alpha=0.1,
    # label="$\pm1\sigma$"
)


plt.errorbar(
    lasf/1e12, lasv, yerr = lasdv,
    ecolor="gray", color='red', label='laser',
    linewidth = 0, elinewidth = 1, markersize = 6, marker='^'
)
plt.plot(fspace/1e12, planck(fspace, lash, lasf0), color='red', lw=1,
                 label=f'$f_0 = {lasf0*1e-12:.0f} \pm {lasdf*1e-12:.0f}$ THz,\n $h = ({lash*1e15:.1f} \pm {lasdh*1e15:.1f})\cdot10^{{-15}}$ eV/Hz')
las0 = planck(lasf, lash, lasf0)
las1 = planck(lasf, lash - lasdh, lasf0 + lasdf)
las2 = planck(lasf, lash + lasdh, lasf0 - lasdf)
ax.fill_between(
    fspace/1e12, 
    planck(fspace, lash - lasdh, lasf0 + lasdf),
    planck(fspace, lash + lasdh, lasf0 - lasdf),
    color='red', alpha=0.1,
    # label="$\pm1\sigma$"
)

plt.grid()
plt.ylim(0, 2)
plt.ylabel('Max kinetic energy (eV)')
plt.legend()
plt.sca(ax_res)

ledchi2 = np.sum((ledv - led0)**2/leddv**2)
laschi2 = np.sum((lasv - las0)**2/lasdv**2)
print(ledchi2, laschi2)

# plt.scatter(ledf/1e12, (ledv - led0)/(led2 - led1))
plt.errorbar(
    ledf/1e12, (ledv - led0)/(led2 - led1), yerr = leddv/(led2 - led1),
    ecolor="gray", color='blue',
    linewidth = 0, elinewidth = 1, markersize = 4, marker='s',
    label=f'$\\tilde\chi^2 = {ledchi2:.1f}$'
)

plt.errorbar(
    lasf/1e12, (lasv - las0)/(las2 - las1), yerr = lasdv/(las2 - las1),
    ecolor="gray", color='red', 
    linewidth = 0, elinewidth = 1, markersize = 6, marker='^',
    label=f'$\\tilde\chi^2 = {laschi2:.1f}$'
)

plt.xlabel('Frequency (THz)')
plt.ylabel('Residual ($\sigma$)')
plt.ylim(-2,2)
plt.legend()
plt.grid()
plt.xlim(400, 800)
plt.axhline(color='black')
plt.tight_layout()
```

```python
'''This converts a given wavelength of light to an 
approximate RGB color value. The wavelength must be given
in nanometers in the range from 380 nm through 750 nm
(789 THz through 400 THz).

Original code by Dan Bruton
http://www.physics.sfasu.edu/astro/color/spectra.html

Python code by https://gist.github.com/error454/65d7f392e1acd4a782fc
Modified for Numpy by ChatGPT
'''

def wavelength_to_rgb(wavelengths, gamma=0.8):
    wavelengths = np.asarray(wavelengths, dtype=float)
    R = np.zeros_like(wavelengths)
    G = np.zeros_like(wavelengths)
    B = np.zeros_like(wavelengths)
    
    # Define conditions
    cond1 = (wavelengths >= 380) & (wavelengths <= 440)
    cond2 = (wavelengths > 440) & (wavelengths <= 490)
    cond3 = (wavelengths > 490) & (wavelengths <= 510)
    cond4 = (wavelengths > 510) & (wavelengths <= 580)
    cond5 = (wavelengths > 580) & (wavelengths <= 645)
    cond6 = (wavelengths > 645) & (wavelengths <= 750)
    
    # Apply calculations based on conditions
    attenuation = 0.3 + 0.7 * (wavelengths - 380) / (440 - 380)
    R[cond1] = ((-(wavelengths[cond1] - 440) / (440 - 380)) * attenuation[cond1]) ** gamma
    B[cond1] = (1.0 * attenuation[cond1]) ** gamma
    
    G[cond2] = ((wavelengths[cond2] - 440) / (490 - 440)) ** gamma
    B[cond2] = 1.0
    
    G[cond3] = 1.0
    B[cond3] = (-(wavelengths[cond3] - 510) / (510 - 490)) ** gamma
    
    R[cond4] = ((wavelengths[cond4] - 510) / (580 - 510)) ** gamma
    G[cond4] = 1.0
    
    R[cond5] = 1.0
    G[cond5] = (-(wavelengths[cond5] - 645) / (645 - 580)) ** gamma
    
    attenuation = 0.3 + 0.7 * (750 - wavelengths) / (750 - 645)
    R[cond6] = (1.0 * attenuation[cond6]) ** gamma

    
    # Stack into an array of RGB tuples
    return np.stack([R, G, B], axis=-1)
```
