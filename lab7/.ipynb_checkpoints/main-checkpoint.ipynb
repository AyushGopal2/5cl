```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
plt.rcParams['font.family'] = 'FreeSerif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['axes.axisbelow'] = True
```

```python
data = [ {
    'm': 100, # mass, grams, positive for pushing, negative for pulling
    't': np.array([4.1,11.9,23.3,38.4,53.8,69.7]), # seconds,
    'Vi': 73, # initial volume, cm3
    'V': np.array([70,69,68,67,66,65]), # volume in piston, cm3
}, {
    'm': 200,
    't': np.array([7,12.6,22.4,29.3,39.8,48.4]),
    'Vi': 72,
    'V': np.array([66,64,63,62,61,60]),
}, {
    'm': -200,
    't': np.array([5,14.1,24.1,33.9,42.2,50.9]),
    'Vi': 20,
    'V': np.array([25,26,27,28,29,30]),
}, {
    'm': -100,
    't': np.array([3.7,9.8,22.2,32.4,44.7,58.1]),
    'Vi': 19,
    'V': np.array([22,23,24,25,26,27]),
} ]
V0 = 182.81 + 4.43 - 14.96 # volume outside piston, cm3
T0 = 273.15 # kelvins
T = T0 + 0.3
dT = 1 # kelvins
v0 = lambda T: (111.78) * np.sqrt(T/T0) # flow speed, m/s
m0 = 48.5 # piston mass, grams
dm = 1 # grams
g = 980 # gravitational accel, cm/s2
A0 = np.pi * (32.5/10/2)**2 # piston head area, cm2
P0 = 1013250 # atmos pressure, dynes/cm2
P = lambda m: P0 + (m + np.sign(m)*m0)*g/A0 # pressure from mass, dynes/cm2
P1 = lambda m: P0 + np.sign(m)*m0*g/A0 # pressure without mass, dynes/cm2
```

```python
model = lambda m, Vi, T: lambda t, A: (
    (Vi+V0) * P1(m)/P(m) - t * A * v0(T) * (1 - P0 / P(m)) - V0
)
fit = dict()
for d in data:
    [A], [[ varA ]] = sp.optimize.curve_fit(
        model(d['m'], d['Vi'], T), d['t'], d['V']
    )
    dA = np.sqrt(varA)
    d['A'] = A
    d['dA'] = dA
    print(A, dA)
```

```python
fig, (ax, res_ax) = plt.subplots(
    nrows=2, figsize=(6,4), dpi=200, 
    sharex=True, gridspec_kw={'height_ratios': [3, 1]}
)

tspace = np.linspace(0, 80, 100)
plt.xlim(0,80)

for d in data:
    m, Vi, t, V, A, dA = d['m'], d['Vi'], d['t'], d['V'], d['A'], d['dA']

    mod = model(m, Vi, T)

    Vspace = mod(tspace, A)
    dVspace = 0.5 * np.abs( mod(tspace, A + dA) - mod(tspace, A - dA) )
    
    Vpred = mod(d['t'], d['A'])
    dVpred = 0.5 * np.abs( mod(t, A + dA) - mod(t, A - dA) )
    Vres = V - Vpred

    rchi2 = np.mean((Vres/dVpred)**2)
    
    marker = '^' if m > 0 else 'v'
    label = f"{np.abs(m):.1f} g {'push' if m > 0 else 'pull'}: $A = {A*100:.1f}\pm{dA*100:.1f}$ mm$^2$, $\\tilde\chi^2 = {rchi2:.1f}$"
    color = 'royalblue' if np.abs(m) > 150 else 'crimson'
    plt.sca(ax)
    plt.scatter(
        t, V + V0,
        label=label,
        marker=marker,
        color=color
    )
    plt.fill_between(
        tspace, Vspace + dVspace + V0, Vspace - dVspace + V0,
        color=color, alpha=0.5
    )
    plt.sca(res_ax)
    plt.scatter(
        t,  Vres,
        marker=marker,
        color=color
    )
    plt.plot(
        t,  Vres,
        color=color,
        alpha=0.5, ls='dashed'
    )

plt.sca(ax)
plt.ylabel('Volume (mL)')
plt.grid()
plt.legend(loc='center left')
plt.sca(res_ax)
plt.xlabel('Time (s)')
plt.ylabel('Residual (mL)')
plt.axhline(0, color='black')
plt.ylim(-1,1)
plt.grid()
plt.tight_layout()
```

```python
A = np.array([d['A'] for d in data])
dA = np.array([d['dA'] for d in data])
Amean = np.mean(A)
Astd = np.std(A)
f'Hole area: {Amean*100:.2f} +/- {Astd*100:.2f} mm^2'
```

```python
fig, (ax) = plt.subplots(
    figsize=(3,2), dpi=200, 
)
mspace = np.linspace(0.1, 300, 100)
mc = []
for i in range(100):
    Vi_mc = np.random.normal(15.375, 2.50) + V0
    T_mc = np.random.uniform(0.3 + T0, 47.6 + T0)
    t_mc = np.random.normal(30, 10)
    A_mc = np.random.normal(Amean, Astd)
    mod = model(mspace, Vi_mc, T_mc)
    mc.append(mod(0, A_mc) - mod(t_mc, A_mc))
leak_mean = np.mean(mc, axis=0)
leak_std = np.std(mc, axis=0)
plt.plot(mspace, leak_mean)
plt.fill_between(mspace, leak_mean - leak_std, leak_mean + leak_std, alpha=0.5)
plt.xlim(0, 250)
plt.ylim(0,6)
plt.grid()
plt.xlabel('Test mass (g)')
plt.ylabel('Leaked volume (mL)')
```
